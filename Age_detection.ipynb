```python
import gradio as gr
import torch
from PIL import Image
from transformers import ViTFeatureExtractor, ViTForImageClassification
```

```python
model = ViTForImageClassification.from_pretrained('nateraw/vit-age-classifier')
transforms = ViTFeatureExtractor.from_pretrained('nateraw/vit-age-classifier')
```

```python
def classify_age(image):
    inputs = transforms(image, return_tensors='pt')

    output = model(**inputs)
    proba = output.logits.softmax(1)

    preds = torch.argmax(proba, dim=1).item()
    probability = proba[0, preds].item()

    age_ranges = {
        0: '0-2',
        1: '3-9',
        2: '10-19',
        3: '20-29',
        4: '30-39',
        5: '40-49',
        6: '50-59',
        7: '60-69',
        8: 'more than 70',
    }

    age_range = age_ranges[preds]

    return f"Score: {probability:.5f}\nLabel: {age_range}"
```

```python
iface = gr.Interface(
    fn=classify_age,
    inputs="image",
    outputs="text",
    title="Age Classification",
)
iface.launch(share=True)
```
